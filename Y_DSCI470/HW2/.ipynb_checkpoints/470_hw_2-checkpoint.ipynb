```python
#print(df)

x = df.iloc[:, :-1]
y = df.iloc[:,-1]

#Get all data into numerical data in order to run coorelation on it
encodedData = pd.get_dummies(df)

# Coorelation is strong metric to identify how good something is at identifying another thing absolute value is used since strong negative or positive are equally beneficial to me
coorelation = encodedData.corr().loc['credit_risk'].abs()

#Since we dont want to consider the coorelation between credit risk and itself
coorelation = coorelation.drop('credit_risk')


#Sort values to make to get highly desired values at top
coorelation = coorelation.sort_values(ascending=False)

#Removes all columns below threshold of .1 to remove noise
for i in range(len(coorelation)):
    if (coorelation.iloc[i] < .125):
        break;

coorelation = coorelation.head(i)



topFeatures = coorelation.index.tolist()

print(topFeatures)
#print(coorelation)


plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='duration', y='credit_amt', hue='credit_risk', palette={1: 'green', 2: 'red'})
#print(y)

plt.show()
```

## Baseline Model ##

Now based on your exploration you should have noted that some features are categorical and some are numerical. As we have learned building a model only makes sense if the data is represented in a numerical way. Therfore, to get a baseline create a classification model of your choosing (ie. knn, logistic, SVM, decision tree, etc.) and split the data into training and testing sets using only numeric features. Report the accuracy of your model on the test set.

```python
from sklearn.svm import SVC

#Gets all numerical data
numericData = df.select_dtypes(include=['number'])

#Split Data
numeric_X_train, numeric_X_test, numeric_y_train, numeric_y_test = train_test_split(numericData, y, test_size=0.33, random_state=42)

#Create and train model
clf = SVC()
clf.fit(numeric_X_train, numeric_y_train)

#Report Accuarcy
print(clf.score(numeric_X_test, numeric_y_test))
```

## One-hot encoding ##

Now that we have a baseline for only the muneric data lets start thinking about the categorical data. Convert the categorical features into a numerical format and train your model using these features. Again report the accuracy  of this model and describe any differences that may occur in the two models.

```python
#Encode Data
encodedData = pd.get_dummies(df)

#Split Data
X_train, X_test, y_train, y_test = train_test_split(encodedData, y, test_size=0.33, random_state=42)

#Create and train model
clf = SVC()
clf.fit(X_train, y_train)

#Report Accuarcy
print(clf.score(X_test, y_test))
```

To Do: accuracy and comparison

The accuracy rose from 70.6% to 70.9% which is a slight increase it seems that the added information from the new categorical data was overall not signifcantly more helpful from the data already inside the numerical data.
