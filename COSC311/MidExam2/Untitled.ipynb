```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_breast_cancer
from sklearn import datasets
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report
from sklearn.model_selection import cross_val_score, KFold
from sklearn import tree
from sklearn.svm import SVC
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
knn = KNeighborsClassifier(n_neighbors = 5)
unsplitData = load_breast_cancer()
```

```python
featureTrain, featureTest, classTrain, classTest = train_test_split(unsplitData.data, unsplitData.target, test_size = .40, random_state= 12)
knn.fit(featureTrain, classTrain)
```

```python
predictions = knn.predict(featureTest)
accuracyData = classification_report(classTest, predictions)
print(accuracyData)
```

```python
DTree = tree.DecisionTreeClassifier(criterion = 'entropy', max_depth=15)
DTree.fit(unsplitData.data, unsplitData.target)

predictions = DTree.predict(unsplitData.data)

confMatrix = confusion_matrix(unsplitData.target, predictions)

sb.heatmap(confMatrix, annot=True)

plt.show()
```

```python
svc = SVC(kernel='rbf', C=5, gamma=.05)

k = 5
kf = KFold(n_splits=k, random_state=None)
 
result = cross_val_score(svc , unsplitData.data, unsplitData.target, cv = kf)
 
print("Avg accuracy: {}".format(result.mean()))
```
