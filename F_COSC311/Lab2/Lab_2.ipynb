# Task 4

```python
monthRelativeHumidity = []
monthlyAverageHumdity = []
months = ['June', 'July', 'August', 'September']

for i in bejaia['month'].unique():
    monthRelativeHumidity.append([bejaia[' RH'][j] for j in range(len(bejaia)) if bejaia['month'][j] == i])
    
for i in range(len(monthRelativeHumidity)):
    monthlyAverageHumdity.append(sum(monthRelativeHumidity[i]) / len(monthRelativeHumidity[i]))


plt.xlabel("Month")
plt.ylabel("Average Humidity")
plt.title("Average Relative Humidity by Month")
plt.xticks(bejaia['month'].unique(), months)


plt.plot(bejaia['month'].unique(), monthlyAverageHumdity, color = 'purple', linestyle='solid', marker = 'v')
```

```python
julyTemps = [sidi_Bel["Temperature"][i] for i in range(len(sidi_Bel["Temperature"]))if sidi_Bel['month'][i] == 6]
julyHumid = [sidi_Bel[" RH"][i] for i in range(len(sidi_Bel["Temperature"]))if sidi_Bel['month'][i] == 6]
coorelation = [i for i in zip(julyHumid, julyTemps)]
coorelation = sorted(coorelation)
julyHumid = [i for i, j in coorelation]
julyTemps = [j for i, j in coorelation]

plt.xlabel("Humidity")
plt.ylabel("Temperature")
plt.title("Coorelation Between Temperature and Relative Humidity")

plt.plot(julyHumid, julyTemps, color = 'black', linestyle='solid', marker = '.')
```

# Task 8

```python
humidityCounts = Counter([i // 10 * 10 for i in bejaia[' RH']])
humidity  = []
numofDays= []

for i in sorted(humidityCounts):
    humidity.append(i)
    numofDays.append(humidityCounts[i])

plt.bar(humidity, numofDays, width = 7, color = 'red')

plt.title("Relative Humdity by Decile")
plt.xlabel("Relative Humidity")
plt.ylabel("Number of Days")

plt.plot()
```

# Task 9

```python
noFireTemp = []
fireTemp = []
#months = ['January', 'Febuary', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

for i in bejaia['month'].unique():
    noFireTemp.append([bejaia['Temperature'][j] for j in range(len(bejaia)) if bejaia['month'][j] == i and bejaia['Classes  '][j].strip().lower() == 'not fire'])
    fireTemp.append([bejaia['Temperature'][j] for j in range(len(bejaia)) if bejaia['month'][j] == i and bejaia['Classes  '][j].strip().lower() == 'fire'])

width = .4
x = np.asarray(range(min(bejaia['month']), max(bejaia['month']) + 1)) 
noFireAvg = [np.mean(i) for i in noFireTemp]
fireAvg = [np.mean(i) for i in fireTemp]


plt.bar(x , noFireAvg, width, label = "No Fire") 
plt.bar(x + width, fireAvg, width, label = "Fire")


plt.title("Monthly Average Temperature During Fire & No Fire")
plt.xlabel("Month")
plt.ylabel("Average Recorded Temperature")

plt.xticks(x + width / 2, months)

plt.legend()

plt.show()
```
